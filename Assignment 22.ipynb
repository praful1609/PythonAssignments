#### Ans1.
What is the result of the code, and explain?

```python
>>> X = 'iNeuron'
>>> def func():
...     print(X)
>>> func()
```

Result:
```
iNeuron
```

Explanation:
In this code, a global variable `X` is defined with the value `'iNeuron'`. The function `func()` is defined, and when it is called, it prints the value of the global variable `X`, which is `'iNeuron'`.

What does this code print, and why?

```python
>>> X = 'iNeuron'
>>> def func():
...     X = 'NI'
...     print(X)
>>> func()
>>> print(X)
```

Result:
```
NI
iNeuron
```

Explanation:
In this code, a global variable `X` is defined with the value `'iNeuron'`. The function `func()` is defined, and within the function, a local variable `X` is defined with the value `'NI'`. When we call the function, it prints the value of the local variable `X`, which is `'NI'`. After the function call, when we print the value of `X` again, it refers to the global variable `X`, which remains `'iNeuron'`.

#### Ans4.
What output does this code produce? Why?

```python
>>> X = 'iNeuron'
>>> def func():
...     global X
...     X = 'NI'
>>> func()
>>> print(X)
```

Result:
```
NI
```

Explanation:
In this code, a global variable `X` is defined with the value `'iNeuron'`. The function `func()` is defined, and within the function, the `global` keyword is used to indicate that we are referring to the global variable `X`. So, when we call the function, it modifies the value of the global variable `X` to `'NI'`. When we print the value of `X` after calling the function, it now reflects the updated value `'NI'`.

What about this code—what’s the output, and why?

```python
>>> X = 'iNeuron'
>>> def func():
...     X = 'NI'
...     def nested():
...         print(X)
...     nested()
>>> func()
>>> X
```

Result:
```
NI
'iNeuron'
```

Explanation:
In this code, a global variable `X` is defined with the value `'iNeuron'`. The function `func()` is defined, and within the function, a local variable `X` is defined with the value `'NI'`. Inside `func()`, another nested function `nested()` is defined, which prints the value of the local variable `X`. When we call `func()`, it executes `nested()`, which prints the value `'NI'`. After the function call, when we print the value of the global variable `X`, it remains `'iNeuron'`.

#### Ans6.
How about this code: what is its output in Python 3, and explain?

```python
>>> def func():
...     X = 'NI'
...     def nested():
...         nonlocal X
...         X = 'Spam'
...     nested()
...     print(X)
>>> func()
```

Result:
```
Spam
```

Explanation:
In this code, the function `func()` is defined. Inside `func()`, there is a local variable `X` with the value `'NI'`. Inside `func()`, another nested function `nested()` is defined. The `nonlocal` keyword is used to indicate that we are referring to the local variable `X` of the enclosing scope (which is `func()`). When we call `nested()`, it modifies the value of the local variable `X` to `'Spam'`. After `nested()` is called, when we print the value of `X` within `func()`, it reflects the updated value `'Spam'`.
