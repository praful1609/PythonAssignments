#### Ans 1.
Python program to extract unique values from dictionary values:

```python
def extract_unique_values(dictionary):
    unique_values = set()
    for value in dictionary.values():
        if isinstance(value, list):
            unique_values.update(value)
        else:
            unique_values.add(value)
    return list(unique_values)

# Test the function
sample_dict = {
    'a': [1, 2, 3],
    'b': [3, 4, 5],
    'c': 5,
    'd': 6,
    'e': 4
}
print("Unique values from dictionary values:", extract_unique_values(sample_dict))
```

#### Ans 2.
Python program to find the sum of all items in a dictionary:

```python
def sum_of_items(dictionary):
    return sum(dictionary.values())

# Test the function
sample_dict = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print("Sum of all items in the dictionary:", sum_of_items(sample_dict))
```

#### Ans 3.
Python program to merge two dictionaries:

```python
def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()
    merged_dict.update(dict2)
    return merged_dict

# Test the function
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
print("Merged dictionary:", merge_dictionaries(dict1, dict2))
```

#### Ans 4.
Python program to convert key-values list to a flat dictionary:

```python
def list_to_flat_dictionary(keys, values):
    return dict(zip(keys, values))

# Test the function
keys = ['a', 'b', 'c']
values = [1, 2, 3]
print("Flat dictionary:", list_to_flat_dictionary(keys, values))
```

#### Ans 5.
Python program to insert at the beginning in OrderedDict:

```python
from collections import OrderedDict

def insert_at_beginning(ordered_dict, key, value):
    ordered_dict.move_to_end(key, last=False)
    ordered_dict[key] = value

# Test the function
ordered_dict = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
print("Original OrderedDict:", ordered_dict)
insert_at_beginning(ordered_dict, 'd', 4)
print("OrderedDict after insertion at the beginning:", ordered_dict)
```

Python program to check the order of characters in a string using OrderedDict:

```python
from collections import OrderedDict

def check_order_of_characters(input_string, pattern):
    char_dict = OrderedDict.fromkeys(pattern)
    j = 0
    for char in input_string:
        if j == len(pattern):
            break
        if char == pattern[j]:
            j += 1
    return j == len(pattern)

# Test the function
input_string = "hello world"
pattern = "hlo"
if check_order_of_characters(input_string, pattern):
    print("The characters in the pattern appear in order in the string.")
else:
    print("The characters in the pattern do not appear in order in the string.")
```

#### Ans 7.
Python program to sort Python Dictionaries by Key or Value:

```python
def sort_dictionary_by_key(dictionary):
    return dict(sorted(dictionary.items()))

def sort_dictionary_by_value(dictionary):
    return dict(sorted(dictionary.items(), key=lambda item: item[1]))

# Test the functions
sample_dict = {'a': 3, 'b': 1, 'c': 2}
print("Dictionary sorted by key:", sort_dictionary_by_key(sample_dict))
print("Dictionary sorted by value:", sort_dictionary_by_value(sample_dict))
```
