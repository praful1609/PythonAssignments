#### Ans 1.
Python program to find words which are greater than a given length k:

```python
def find_words_greater_than_k(sentence, k):
    words = sentence.split()
    return [word for word in words if len(word) > k]

# Test the function
sentence = "This is a sample sentence to find words greater than k."
k = 4
print("Words greater than length", k, ":", find_words_greater_than_k(sentence, k))
```

#### Ans 2.
Python program for removing the i-th character from a string:

```python
def remove_ith_character(input_string, i):
    return input_string[:i] + input_string[i+1:]

# Test the function
input_string = "Hello, World!"
i = 4
print("String after removing the", i, "th character:", remove_ith_character(input_string, i))
```

#### Ans 3.
Python program to split and join a string:

```python
def split_and_join_string(input_string):
    words = input_string.split()
    return "-".join(words)

# Test the function
input_string = "This is a sample string."
print("String after splitting and joining:", split_and_join_string(input_string))
```

#### Ans 4.
Python program to check if a given string is a binary string or not:

```python
def is_binary_string(input_string):
    return all(char in "01" for char in input_string)

# Test the function
input_string = "101010"
if is_binary_string(input_string):
    print(input_string, "is a binary string.")
else:
    print(input_string, "is not a binary string.")
```

#### Ans 5.
Python program to find uncommon words from two strings:

```python
def find_uncommon_words(str1, str2):
    set1 = set(str1.split())
    set2 = set(str2.split())
    return list(set1.symmetric_difference(set2))

# Test the function
str1 = "This is a sample string 1."
str2 = "This is a sample string 2."
print("Uncommon words:", find_uncommon_words(str1, str2))
```

#### Ans 6.
Python program to find all duplicate characters in a string:

```python
def find_duplicate_characters(input_string):
    duplicate_chars = [char for char in input_string if input_string.count(char) > 1]
    return list(set(duplicate_chars))

# Test the function
input_string = "Hello, World!"
print("Duplicate characters:", find_duplicate_characters(input_string))
```

#### Ans 7.
Python program to check if a string contains any special character:

```python
import re

def has_special_characters(input_string):
    return bool(re.search("[^a-zA-Z0-9]", input_string))

# Test the function
input_string = "Hello123!"
if has_special_characters(input_string):
    print(input_string, "contains special characters.")
else:
    print(input_string, "does not contain special characters.")
```
