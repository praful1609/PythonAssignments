To call `logging.debug()`, you need to include the following two lines at the beginning of your Python script:


import logging
logging.basicConfig(level=logging.DEBUG)


#### Ans5.
To send `logging.debug()` messages to a file named "programLog.txt," you need to add the following two lines along with the previous ones:


import logging
logging.basicConfig(filename='programLog.txt', level=logging.DEBUG)


#### Ans6.
The five levels of logging, in increasing order of severity, are:

   - DEBUG: Detailed information, typically useful for diagnosing problems.
   - INFO: Confirmation that things are working as expected.
   - WARNING: An indication that something unexpected or potentially harmful happened.
   - ERROR: An error that caused a function or operation to fail.
   - CRITICAL: A critical error that might lead to the termination of the program.

#### Ans7.
To disable all logging messages, you can add the following line of code at the beginning of your script:


logging.disable(logging.CRITICAL)


#### Ans8.
Using logging messages is better than using print() for several reasons:
   - Logging allows you to control the level of messages displayed (e.g., DEBUG, INFO, WARNING) and easily filter them.
   - Logging provides timestamps and additional information about each log entry.
   - Logging messages can be easily directed to different destinations (e.g., console, file) without modifying the code.
   - Logging can be enabled or disabled globally, making it easy to manage log output during development and production.

#### Ans9.
In the debugger:

   - Step Over: Executes the current line of code and advances to the next line in the current function. If the current line contains a function call, it will not step into that function but instead execute it and move to the next line after the function call.
   - Step In: If the current line contains a function call, it will step into that function and pause at the first line of the called function.
   - Step Out: Continues execution until the current function returns and pauses on the line that called the current function.

#### Ans10.
After you click Continue in the debugger, the debugger will stop when it reaches the next breakpoint (if any) or when the program execution completes.

#### Ans11.
A breakpoint is a specific point in the code where the debugger will pause program execution, allowing you to inspect variables, step through the code, and analyze the program's state at that point. It is a tool used for debugging and helps in understanding how the program behaves at specific locations. You can set breakpoints in your code to stop the execution and examine the program's state at that particular moment.
