#### Question 1:

```python
def stutter(word):
    if len(word) >= 2:
        repeated_part = word[:2] + "... "
        return repeated_part + repeated_part + word + "?"
    return "Word should be at least 2 characters long."

# Test the function
print(stutter("incredible"))
print(stutter("enthusiastic"))
print(stutter("outstanding"))
```

#### Question 2:

```python
import math

def radians_to_degrees(radians):
    return round(math.degrees(radians), 1)

# Test the function
print(radians_to_degrees(1))
print(radians_to_degrees(20))
print(radians_to_degrees(50))
```

#### Question 3:

```python
def is_curzon(num):
    numerator = 2 ** num + 1
    denominator = 2 * num + 1
    return numerator % denominator == 0

# Test the function
print(is_curzon(5))
print(is_curzon(10))
print(is_curzon(14))
```

#### Question 4:

```python
import math

def area_of_hexagon(x):
    return round((3 * math.sqrt(3) * x ** 2) / 2, 1)

# Test the function
print(area_of_hexagon(1))
print(area_of_hexagon(2))
print(area_of_hexagon(3))
```

#### Question 5:

```python
def binary(decimal_number):
    return bin(decimal_number)[2:]

# Test the function
print(binary(1))
print(binary(5))
print(binary(10))
```
