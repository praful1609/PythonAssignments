#### Ans1.
The `PdfFileReader()` and `PdfFileWriter()` File objects should be opened in binary mode ('rb') when working with PDF files.

#### Ans2.
To get a Page object for page 5 from a `PdfFileReader` object (let's say `pdf_reader`), you can use the `getPage()` method, like this:

```python
page_number = 5  # Page numbers are 0-indexed
page = pdf_reader.getPage(page_number)
```

#### Ans3.
The `PdfFileReader` variable that stores the number of pages in the PDF document is `numPages`. You can access it like this:

```python
num_pages = pdf_reader.numPages
```

#### Ans4.
If a `PdfFileReader` object's PDF is encrypted with the password 'swordfish', you must use the `decrypt()` method and pass the correct password before obtaining Page objects. Here's how you would do it:

```python
pdf_reader.decrypt('swordfish')
```

#### Ans5.
To rotate a page, you can use the `rotateClockwise()` and `rotateCounterClockwise()` methods of the `Page` object. For example, to rotate the page 90 degrees clockwise, you can do:

```python
page.rotateClockwise(90)
```

#### Ans6.
In the context of working with python-docx library (for Word documents):
   - A `Run` object represents a contiguous run of text with the same character style in a paragraph.
   - A `Paragraph` object represents a single paragraph of text.

#### Ans7.
To obtain a list of `Paragraph` objects for a `Document` object stored in a variable named `doc`, you can use the `paragraphs` attribute of the `Document` object. For example:

```python
paragraphs_list = doc.paragraphs
```

#### Ans8.
The type of object that has bold, underline, italic, strike, and outline variables is the `Run` object. A `Run` object represents a contiguous run of text with the same character style within a paragraph.

#### Ans9.
For the `bold` variable in a `Run` object:
   - `True`: The text in the run is formatted as bold.
   - `False`: The text in the run is not bold.
   - `None`: The text in the run may have a mix of bold and non-bold formatting or inherits its bold formatting from the paragraph or document style.

#### Ans10.
To create a `Document` object for a new Word document, you need to import the `Document` class from the python-docx library and create a new instance of it:

```python
from docx import Document

doc = Document()
```

#### Ans11.
To add a paragraph with the text 'Hello, there!' to a `Document` object stored in a variable named `doc`, you can use the `add_paragraph()` method of the `Document` object:

```python
doc.add_paragraph('Hello, there!')
```
