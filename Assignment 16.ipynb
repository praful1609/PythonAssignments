#### Ans1.
Create a list called `years_list` starting with the year of your birth and each year thereafter until the year of your fifth birthday:

```python
years_list = [year for year in range(1980, 1980 + 6)]
print(years_list)  # Output: [1980, 1981, 1982, 1983, 1984, 1985]
```

#### Ans2.
To find the year in `years_list` when you had your third birthday, you can simply access the element at index 3 (remembering that the first year is at index 0):

```python
third_birthday_year = years_list[3]
print(third_birthday_year)  # Output: 1983
```

#### Ans3.
To find the year when you were the oldest in `years_list`, you can use the `max()` function:

```python
oldest_year = max(years_list)
print(oldest_year)  # Output: 1985
```

#### Ans6.
Make a surprise list with the elements "Groucho", "Chico", and "Harpo":

```python
surprise = ["Groucho", "Chico", "Harpo"]
```

#### Ans7.
To lowercase the last element of the surprise list, reverse it, and then capitalize it:

```python
surprise[-1] = surprise[-1].lower()[::-1].capitalize()
print(surprise)  # Output: ['Groucho', 'Chico', 'Oprah']
```

#### Ans8.
Make an English-to-French dictionary called `e2f` and print it:

```python
e2f = {'dog': 'chien', 'cat': 'chat', 'walrus': 'morse'}
print(e2f)
```

#### Ans9.
To write the French word for walrus in your dictionary `e2f`:

```python
french_word_for_walrus = e2f['walrus']
print(french_word_for_walrus)  # Output: 'morse'
```

#### Ans10.
Make a French-to-English dictionary called `f2e` from `e2f` using the `items()` method:

```python
f2e = {french: english for english, french in e2f.items()}
print(f2e)
```

#### Ans11.
To print the English version of the French word "chien" using `f2e`:

```python
english_word_for_chien = f2e['chien']
print(english_word_for_chien)  # Output: 'dog'
```

#### Ans12.
To make and print a set of English words from the keys in `e2f`:

```python
english_words_set = set(e2f.keys())
print(english_words_set)
```

#### Ans13.
Make a multilevel dictionary called `life` with the given structure:

```python
life = {
    'animals': {
        'cats': ['Henri', 'Grumpy', 'Lucy'],
        'octopi': {},
        'emus': {}
    },
    'plants': {},
    'other': {}
}
```

#### Ans14.
To print the top-level keys of `life`:

```python
top_level_keys = life.keys()
print(top_level_keys)
```

#### Ans15.
To print the keys for `life['animals']`:

```python
animal_keys = life['animals'].keys()
print(animal_keys)
```
