#### Ans 1.
Python program to check if a number is Positive, Negative, or Zero:

```python
def check_number(n):
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"

# Test the function
num = float(input("Enter a number: "))
print("The number is", check_number(num))
```

#### Ans 2.
Python program to check if a number is Odd or Even:

```python
def check_odd_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Test the function
num = int(input("Enter a number: "))
print("The number is", check_odd_even(num))
```

#### Ans 3.
Python program to check Leap Year:

```python
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

# Test the function
year = int(input("Enter a year: "))
if is_leap_year(year):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

#### Ans 4.
Python program to check Prime Number:

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test the function
num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

#### Ans 5.
Python program to print all Prime Numbers in an Interval of 1-10000:

```python
def print_primes_in_interval(start, end):
    primes = []
    for num in range(start, end + 1):
        if is_prime(num):
            primes.append(num)
    return primes

# Test the function for the interval [1, 10000]
start = 1
end = 10000
print("Prime numbers in the interval [1, 10000]:")
print(print_primes_in_interval(start, end))
```
