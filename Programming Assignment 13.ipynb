#### Question 1:

```python
import math

def calculate_q(sequence):
    C, H = 50, 30
    result = []
    for D in sequence:
        Q = int(math.sqrt((2 * C * int(D)) / H))
        result.append(Q)
    return result

# Test the function
sequence = input("Enter comma-separated values for D: ").split(',')
output = calculate_q(sequence)
print(','.join(map(str, output)))
```

#### Question 2:

```python
def generate_2D_array(X, Y):
    return [[i*j for j in range(Y)] for i in range(X)]

# Test the function
X, Y = map(int, input("Enter 2 digits for X and Y: ").split(','))
output = generate_2D_array(X, Y)
for row in output:
    print(row)
```

#### Question 3:

```python
def sort_words_alphabetically(words):
    words = words.split(',')
    words.sort()
    return ','.join(words)

# Test the function
words_input = input("Enter comma-separated words: ")
output = sort_words_alphabetically(words_input)
print(output)
```

#### Question 4:

```python
def remove_duplicates_and_sort(sentence):
    words = sentence.split()
    words = sorted(set(words))
    return ' '.join(words)

# Test the function
sentence_input = input("Enter whitespace-separated words: ")
output = remove_duplicates_and_sort(sentence_input)
print(output)
```

#### Question 5:

```python
def count_letters_and_digits(sentence):
    letters = sum(1 for char in sentence if char.isalpha())
    digits = sum(1 for char in sentence if char.isdigit())
    return letters, digits

# Test the function
sentence_input = input("Enter a sentence: ")
letters, digits = count_letters_and_digits(sentence_input)
print("LETTERS", letters)
print("DIGITS", digits)
```

#### Question 6:

```python
import re

def check_valid_password(passwords):
    valid_passwords = []
    for password in passwords.split(','):
        if 6 <= len(password) <= 12 and re.search("[a-z]", password) and re.search("[A-Z]", password) and re.search("[0-9]", password) and re.search("[$#@]", password):
            valid_passwords.append(password)
    return ','.join(valid_passwords)

# Test the function
passwords_input = input("Enter comma-separated passwords: ")
output = check_valid_password(passwords_input)
print(output)
```
