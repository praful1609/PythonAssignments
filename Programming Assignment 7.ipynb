#### Ans 1.
Python program to find the sum of an array:

```python
def array_sum(arr):
    return sum(arr)

# Test the function
arr = [1, 2, 3, 4, 5]
print("Sum of the array:", array_sum(arr))
```

#### Ans 2.
Python program to find the largest element in an array:

```python
def find_largest_element(arr):
    return max(arr)

# Test the function
arr = [15, 6, 27, 8, 19]
print("Largest element in the array:", find_largest_element(arr))
```

#### Ans 3.
Python program for array rotation:

```python
def rotate_array(arr, k):
    k = k % len(arr)
    return arr[k:] + arr[:k]

# Test the function
arr = [1, 2, 3, 4, 5]
k = 2
print("Original array:", arr)
print("Rotated array:", rotate_array(arr, k))
```

#### Ans 4.
Python program to split the array and add the first part to the end:

```python
def split_and_add(arr, split_index):
    return arr[split_index:] + arr[:split_index]

# Test the function
arr = [1, 2, 3, 4, 5]
split_index = 2
print("Original array:", arr)
print("Split and added array:", split_and_add(arr, split_index))
```

#### Ans 5.
Python program to check if the given array is Monotonic:

```python
def is_monotonic(arr):
    increasing = decreasing = True

    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            increasing = False
        if arr[i] > arr[i - 1]:
            decreasing = False

    return increasing or decreasing

# Test the function
arr = [1, 2, 3, 4, 5]
print("Is the array monotonic?", is_monotonic(arr))
```
