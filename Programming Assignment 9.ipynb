#### Ans 1.
Python program to check if the given number is a Disarium Number:

```python
def is_disarium_number(num):
    num_str = str(num)
    sum_digits = sum(int(digit) ** (index + 1) for index, digit in enumerate(num_str))
    return num == sum_digits

# Test the function
num = int(input("Enter a number: "))
if is_disarium_number(num):
    print(num, "is a Disarium number.")
else:
    print(num, "is not a Disarium number.")
```

#### Ans 2.
Python program to print all Disarium numbers between 1 to 100:

```python
def find_disarium_numbers(start, end):
    disarium_numbers = []
    for num in range(start, end + 1):
        if is_disarium_number(num):
            disarium_numbers.append(num)
    return disarium_numbers

# Test the function for the interval [1, 100]
start = 1
end = 100
print("Disarium numbers in the interval [1, 100]:")
print(find_disarium_numbers(start, end))
```

#### Ans 3.
Python program to check if the given number is a Happy Number:

```python
def is_happy_number(num):
    seen = set()
    while num != 1 and num not in seen:
        seen.add(num)
        num = sum(int(digit) ** 2 for digit in str(num))
    return num == 1

# Test the function
num = int(input("Enter a number: "))
if is_happy_number(num):
    print(num, "is a Happy number.")
else:
    print(num, "is not a Happy number.")
```

#### Ans 5.
Python program to determine whether the given number is a Harshad Number:

```python
def is_harshad_number(num):
    return num % sum(int(digit) for digit in str(num)) == 0

# Test the function
num = int(input("Enter a number: "))
if is_harshad_number(num):
    print(num, "is a Harshad number.")
else:
    print(num, "is not a Harshad number.")
```

6. Python program to print all Pronic numbers between 1 and 100:

```python
def find_pronic_numbers(start, end):
    pronic_numbers = []
    for num in range(start, end + 1):
        if num == (num ** 0.5) * (num ** 0.5 + 1):
            pronic_numbers.append(num)
    return pronic_numbers

# Test the function for the interval [1, 100]
start = 1
end = 100
print("Pronic numbers in the interval [1, 100]:")
print(find_pronic_numbers(start, end))
```
