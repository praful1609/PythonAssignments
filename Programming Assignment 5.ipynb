#### Ans 1.
Python program to find LCM (Least Common Multiple):

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    return (a * b) // gcd(a, b)

# Test the function
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
print("LCM of", num1, "and", num2, "is", lcm(num1, num2))
```

#### Ans 2.
Python program to find HCF (Highest Common Factor):

```python
def hcf(a, b):
    while b:
        a, b = b, a % b
    return a

# Test the function
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
print("HCF of", num1, "and", num2, "is", hcf(num1, num2))
```

#### Ans 3.
Python program to convert Decimal to Binary, Octal, and Hexadecimal:

```python
def decimal_to_binary(decimal_num):
    return bin(decimal_num)

def decimal_to_octal(decimal_num):
    return oct(decimal_num)

def decimal_to_hexadecimal(decimal_num):
    return hex(decimal_num)

# Test the functions
decimal_num = int(input("Enter a decimal number: "))
print("Binary:", decimal_to_binary(decimal_num))
print("Octal:", decimal_to_octal(decimal_num))
print("Hexadecimal:", decimal_to_hexadecimal(decimal_num))
```

#### Ans 4.
Python program to find ASCII value of a character:

```python
def char_to_ascii(char):
    return ord(char)

# Test the function
char = input("Enter a character: ")
print("ASCII value of", char, "is", char_to_ascii(char))
```

#### Ans 5.
Python program to make a simple calculator with 4 basic mathematical operations:

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        raise ValueError("Division by zero is not allowed.")
    return a / b

# Test the calculator
num1 = float(input("Enter the first number: "))
operator = input("Enter the operator (+, -, *, /): ")
num2 = float(input("Enter the second number: "))

if operator == '+':
    result = add(num1, num2)
elif operator == '-':
    result = subtract(num1, num2)
elif operator == '*':
    result = multiply(num1, num2)
elif operator == '/':
    result = divide(num1, num2)
else:
    print("Invalid operator.")
    exit()

print("Result:", result)
```
